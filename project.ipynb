```python
import random
import SimpleGUICS2Pygame.simpleguics2pygame as simplegui

# Constants
WIDTH = 400
HEIGHT = 400
GRID_SIZE = 4
LINE_WIDTH = 5
TILE_SIZE = WIDTH // GRID_SIZE
TILE_SPEED = 500  # Reduced time interval in milliseconds for tile appearance
WHITE = 'white'
RED = 'red'

# Global variables
tiles = set()  # Use a set to store white tiles as (row, col) strings
score = 0
game_active = False
game_state = "Start New Game"
count = True  # Game state variable
game_timer = 0
# Timer handler to create white tiles


def create_white_tile():
    if game_active:
        row, col = random.randint(
            0, GRID_SIZE - 1), random.randint(0, GRID_SIZE - 1)
        tiles.add(f"{row},{col}")  # Store (row, col) as a string

# Mouse click handler


def mouse_click(pos):
    global score, game_active
    if game_active:
        row, col = pos[1] // TILE_SIZE, pos[0] // TILE_SIZE
        tile_str = f"{row},{col}"  # Convert (row, col) to a string

        if tile_str in tiles:
            tiles.remove(tile_str)
            score += 1
            update_score_label()
        else:
            game_active = False
            game_timer_clock.stop()  # End the game if the user clicks on the white canvas

# Draw handler


def draw(canvas):
    # Draw grid lines (horizontal and vertical)

    if not game_active:
        if count:
            canvas.draw_text("Start New Game", (70,
                                                200), 40, 'Red')
        else:
            canvas.draw_text("Game Over", (110,
                                           150), 40, 'Red')
            canvas.draw_text("Score: " + str(score),
                             (150, 260), 30, 'Black')
    else:
        for i in range(1, GRID_SIZE):
            position = i * TILE_SIZE
            canvas.draw_line(
                (position, 0), (position, HEIGHT), LINE_WIDTH, RED)
            canvas.draw_line(
                (0, position), (WIDTH, position), LINE_WIDTH, RED)

    # Draw white tiles
        for tile_str in tiles:
            row, col = map(int, tile_str.split(','))
            x0, y0 = col * TILE_SIZE, row * TILE_SIZE
            x1, y1 = x0 + TILE_SIZE, y0
            x2, y2 = x1, y0 + TILE_SIZE
            x3, y3 = x0, y2
            canvas.draw_polygon([(x0, y0), (x1, y1), (x2, y2),
                                (x3, y3)], 1, 'Black', 'black')


def update_score_label():
    score_label.set_text("Score: " + str(score))

# Reset game handler


def reset_game():
    global score, game_active, game_state, count
    score = 0
    game_active = True
    game_state = 'Playing'
    tiles.clear()
    update_score_label()
    timer.start()
    count = False
    game_timer_clock.start()


def update_timer_label():
    Timer_label.set_text('Timer : '+str(game_timer)+' (seconds)')


def GameTimerHandler():
    global game_timer, game_active

    if game_timer == 10:
        game_active = False
        game_timer = 0
        game_timer_clock.stop()
    else:
        game_timer += 1
    update_timer_label()


# Create the frame and set handlers
frame = simplegui.create_frame("Don't Tap White Tiles", WIDTH, HEIGHT)
frame.set_canvas_background('White')
frame.set_mouseclick_handler(mouse_click)
score_label = frame.add_label("Score: 0")
frame.add_button(game_state, reset_game)
frame.set_draw_handler(draw)
Timer_label = frame.add_label('Timer : 0 (seconds)')
# Create a timer to generate white tiles
timer = simplegui.create_timer(TILE_SPEED, create_white_tile)
game_timer_clock = simplegui.create_timer(1000, GameTimerHandler)
# Start the game and initial white tiles
frame.start()
create_white_tile()
create_white_tile()
create_white_tile()
create_white_tile()
```
